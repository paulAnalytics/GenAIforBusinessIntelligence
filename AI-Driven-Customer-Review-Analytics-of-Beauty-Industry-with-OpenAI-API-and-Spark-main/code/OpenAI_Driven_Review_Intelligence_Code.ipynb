```
import requests
import json
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import ArrayType, StringType, StructType, StructField
api_key = 'Input Your Key Here'

previous_topics_broadcast = spark.sparkContext.broadcast(set())
import requests
import json

# Define the OpenAI API endpoint and headers
api_endpoint = "https://api.openai.com/v1/engines/text-davinci-003/completions"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

# Modify the UDF to use the new API
@F.udf(ArrayType(StringType()))
def generate_topics_from_batch_udf(batch):
    # Access the broadcast variable for previous topics
    previous_topics = previous_topics_broadcast.value

    # Combine the batch with existing topics
    prompt = f"""For the following reviews: Summarize 5 topics (cannot be more than 4 words) from the reviews below: \n{batch}.
    These are reviews that are covered already {previous_topics}.
    From the 5 summarized topics, Return only topics that are not covered already.  """

    # Prepare the data for the API request
    data = {
        "prompt": prompt,
        "max_tokens": 1500,
        "temperature": 0.000
    }

    # Make the API request
    response = requests.post(api_endpoint, headers=headers, data=json.dumps(data))
    print(response.text)


    # Extract new topics from the response
    new_topics = ["".join(char for char in choice['text'].strip() if not char.isdigit()) for choice in response.json()['choices']]

    return [topic for topic in new_topics if topic not in previous_topics]
```

```
# File location and type
file_location = "/FileStore/tables/Dataset_for_dashboard-1.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
lakme_sampled_data = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(lakme_sampled_data)
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize

# Ensure 'Review' column is of StringType
lakme_sampled_data = lakme_sampled_data.withColumn("Review", lakme_sampled_data["Review"].cast("string"))

# Initialize the VADER sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Define the calculate_relevance function
def calculate_relevance(review):
    relevant_words = ['product', 'quality', 'effective', 'useful', 'value',
                      'texture', 'fragrance', 'long-lasting', 'moisturizing',
                      'pigmentation', 'easy-to-use', 'shade', 'packaging', 'natural',
                      'smooth', 'blendable', 'hydrating', 'matte', 'durable']
    tokens = word_tokenize(review.lower())
    relevant_tokens = [token for token in tokens if token in relevant_words]
    return len(relevant_tokens) / len(tokens) if len(tokens) > 0 else 0

# Define the calculate_coherence function
def calculate_coherence(review):
    sentences = nltk.sent_tokenize(review)

    # Calculate total number of words in all sentences
    total_words = sum(len(word_tokenize(sentence)) for sentence in sentences)

    # Calculate average sentence length and normalize
    average_sentence_length = total_words / len(sentences) if len(sentences) > 0 else 0
    normalized_coherence = min(1, average_sentence_length / 20)
    return normalized_coherence

# Define the calculate_informativeness function
def calculate_informativeness(review):
    tokens = word_tokenize(review.lower())
    unique_tokens = set(tokens)
    return len(unique_tokens) / len(tokens) if len(tokens) > 0 else 0

# Define UDFs for the score calculations
calculate_positivity_udf = udf(lambda review: sia.polarity_scores(review)['compound'], FloatType())
calculate_relevance_udf = udf(lambda review: calculate_relevance(review), FloatType())
calculate_coherence_udf = udf(lambda review: calculate_coherence(review), FloatType())
calculate_informativeness_udf = udf(lambda review: calculate_informativeness(review), FloatType())

# Apply UDFs to calculate scores for lakme_sampled_data
lakme_sampled_data = lakme_sampled_data \
    .withColumn("positivity_score", calculate_positivity_udf("Review")) \
    .withColumn("relevance_score", calculate_relevance_udf("Review")) \
    .withColumn("coherence_score", calculate_coherence_udf("Review")) \
    .withColumn("informativeness_score", calculate_informativeness_udf("Review"))

# Show the result
lakme_sampled_data.show(3)
```
